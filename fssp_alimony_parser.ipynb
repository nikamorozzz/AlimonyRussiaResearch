```python
def pivot(df):
    irow2rowname = {i : j for i, j in df[['Row number', 'Row name']].drop_duplicates().values}
    icol2colname = {i : j for i, j in df[['Column number', 'Column name']].drop_duplicates().values}
    x = df.pivot(index='Row number', columns='Column number', values='Value')
    x.index = [irow2rowname[i] for i in x.index.tolist()]
    x.columns = [icol2colname[i] for i in x.columns.tolist()]
    return x
pivot(razdel_5)
```

```python
import numpy as np

def get_super_index(df, sub_line='в том числе', valid=lambda x : 'алим' in x):
    index = np.array(df.index.tolist())
    super_indexes = np.where([x[:len(sub_line)] != sub_line for x in index])[0].tolist() + [df.shape[0]]
    index2super = {}
    for start, end in zip(super_indexes[:-1], super_indexes[1:]):
        for i in range(start, end):
            index2super[i] = start
    valid_index = np.array([valid(x) for x in index])
    for i in np.where(valid_index)[0]:
        valid_index[index2super[i]] = True
    return valid_index

def get_super_columns(df, sub_line='в том числе', valid=lambda x : 'алим' in x):
    index = np.array(df.columns.tolist())
    super_indexes = np.where([x[:len(sub_line)] != sub_line for x in index])[0].tolist() + [df.shape[1]]
    index2super = {}
    for start, end in zip(super_indexes[:-1], super_indexes[1:]):
        for i in range(start, end):
            index2super[i] = start
    valid_index = np.array([valid(x) for x in index])
    for i in np.where(valid_index)[0]:
        valid_index[index2super[i]] = True
    return valid_index

x = pivot(razdel_5)
alimony_5 = x.loc[:, get_super_columns(x)]
```
